```python
# Future value
pv = 100
r = 0.1
n = 7
fv = pv * ((1+ (r))**n)
print(round(fv, 2))
```

```python
amount = 100
interest = 10 #10% = 0.01 * 10
years = 7

future_value = amount * ((1 + (0.01 * interest)) ** years)
print(round(future_value, 2))
```
