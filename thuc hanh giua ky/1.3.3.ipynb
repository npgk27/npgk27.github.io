```python
import pandas as pd
import sweetviz as sv
import os

# Đảm bảo thư mục lưu báo cáo tồn tại
os.makedirs("Reports", exist_ok=True)

# Đọc dữ liệu (chú ý sep="\t")
data = pd.read_csv("C:/Users/BOOK 1/OneDrive/Desktop/Job/marketing_campaign.csv", sep="\t")

# Kiểm tra sơ bộ dữ liệu
print("Kích thước dữ liệu:", data.shape)
print("Các cột:", data.columns.tolist())
data.head()
```

```python
# Phân tích dữ liệu với biến mục tiêu "Response"
report = sv.analyze(data, target_feat="Response")

# Xuất báo cáo ra file HTML
report.show_html("Reports/Marketing_Sweetviz_Report.html")

print("✅ Báo cáo SweetViz đã được tạo trong thư mục 'Reports'")
```
