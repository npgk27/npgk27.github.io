```python
# 1. Import thư viện
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Nạp dữ liệu
retail = pd.read_excel("C:/Users/BOOK 1/OneDrive/Desktop/Job/Online Retail.xlsx")
```

```python
# 3. Kiểm tra thông tin dữ liệu
retail.info()
retail.describe()
```

```python
# 4. Xử lý dữ liệu thiếu và trùng lặp
retail = retail.dropna()
retail = retail.drop_duplicates()
```

```python
# 5. Tạo biến 'TotalPrice'
retail["TotalPrice"] = retail["Quantity"] * retail["UnitPrice"]
```

```python
# 6. Trực quan hóa top sản phẩm bán chạy nhất
top_products = retail.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 10 sản phẩm bán chạy nhất")
plt.xlabel("Tổng số lượng bán")
plt.ylabel("Tên sản phẩm")
plt.show()
```

```python
# 7. Doanh thu theo quốc gia
country_sales = retail.groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=country_sales.values, y=country_sales.index)
plt.title("Top 10 quốc gia có doanh thu cao nhất")
plt.xlabel("Doanh thu (USD)")
plt.ylabel("Quốc gia")
plt.show()
```

```python
# 8. Phân bố giá trị đơn hàng
plt.figure(figsize=(10,5))
sns.histplot(retail["TotalPrice"], bins=50)
plt.xlim(0,100)
plt.title("Phân bố giá trị đơn hàng (giới hạn <100 USD)")
plt.xlabel("Giá trị đơn hàng")
plt.ylabel("Số lượng đơn")
plt.show()
```
