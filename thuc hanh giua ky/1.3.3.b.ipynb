```python
import pandas as pd
from autoviz import AutoViz_Class

# Đọc dữ liệu
data = pd.read_csv("C:/Users/BOOK 1/OneDrive/Desktop/Job/marketing_campaign.csv", sep="\t")

# Kiểm tra nhanh
print("Kích thước dữ liệu:", data.shape)
data.head()
```

```python
# Khởi tạo AutoViz
AV = AutoViz_Class()

# Tạo báo cáo tự động (EDA)
report = AV.AutoViz(
    "C:/Users/BOOK 1/OneDrive/Desktop/Job/marketing_campaign.csv",  # đường dẫn file
    sep="\t",                # dấu phân tách
    depVar="Response",       # biến mục tiêu
    dfte=None,               # không chia train/test
    header=0, 
    verbose=1,               # in thông tin ra màn hình
    lowess=False, 
    chart_format="svg",      # định dạng đồ thị
    max_rows_analyzed=5000,  # số dòng tối đa để vẽ biểu đồ
    max_cols_analyzed=50     # số cột tối đa
)
```
