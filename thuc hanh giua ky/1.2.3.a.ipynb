```python
# 1. Import thư viện
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Nạp dữ liệu
data = pd.read_csv("C:\\Users\\BOOK 1\\OneDrive\\Desktop\\Job\\diabetes.csv")
data.head()
```

```python
# 3. Thống kê mô tả
data.describe()
```

```python
# 4. Kiểm tra dữ liệu thiếu
data.isnull().sum()
```

```python
# 5. Trực quan hóa phân bố các biến (Histogram)
plt.figure(figsize=(12,6))
sns.histplot(data["BMI"], bins=30, kde=True, color='skyblue')
plt.title("Phân bố chỉ số BMI của bệnh nhân")
plt.xlabel("BMI")
plt.ylabel("Tần suất")
plt.show()
```

```python
# 6. Boxplot để kiểm tra giá trị ngoại lai
plt.figure(figsize=(10,5))
sns.boxplot(x=data["Glucose"])
plt.title("Phân bố Glucose trong máu")
plt.show()
```

```python
# 7. Heatmap thể hiện tương quan giữa các biến
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap='Blues')
plt.title("Ma trận tương quan giữa các biến")
plt.show()
```

```python
# 8. So sánh trung bình BMI theo tình trạng tiểu đường
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x="Outcome", y="BMI")
plt.title("So sánh BMI giữa nhóm mắc và không mắc tiểu đường")
plt.xlabel("Outcome (0 = Không mắc, 1 = Mắc)")
plt.ylabel("BMI")
plt.show()
```
